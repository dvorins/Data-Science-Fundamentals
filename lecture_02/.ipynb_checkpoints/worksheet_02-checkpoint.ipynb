### Effective Programming

a) What is a drawback of the top down approach?

Can cause you to forget little details that may have implications in the chain of operations, (types, memory, helpers, etc.).

Also this pretty much sets the framework in stone, once you create it: there is not much flexibiltiy.

b) What is a drawback of the bottom up approach?

Horse with blinders: unable to see bigger implications of all decisions.

Also can accientatlly create duplicates of functions, or having varying data types for a simliar object.

    1.a search for documentation about error
    1.b try and see if the error has a proposed solution

a) Complete the code below to read in a filepath to the `train.csv` and returns the DataFrame.

```python
import pandas as pd

df = pd.read_csv('/Users/samdvorin/Desktop/code/506/Data-Science-Fundamentals/lecture_02/train.csv')
df.describe()
df.head()
```

b) Complete the code so it returns the number of rows that have at least one empty column value

```python
print("there are " +  str(df[df.isnull().any(axis=1)].shape[0]) + " rows with at least one empty value")
```

c) Complete the code below to remove all columns with more than 200 NaN values

```python
nan_counts = df.isnull().sum()
print(nan_counts)

# Identify columns with more than 200 NaN values
columns_to_drop = nan_counts[nan_counts > 200].index
print(columns_to_drop)

# Drop columns with more than 200 NaN values
df = df.drop(columns=columns_to_drop)

df.columns
```

d) Complete the code below to replaces `male` with 0 and `female` with 1

```python
df['Sex'] = df['Sex'].replace({'male': 0, 'female': 1})
df.head()
```

e) Complete the code below to add four columns `First Name`, `Middle Name`, `Last Name`, and `Title` corresponding to the value in the `name` column.

For example: `Braund, Mr. Owen Harris` would be: 

|First Name | Middle Name | Last Name | Title |
|-----------|-------------|-----------|-------|
| Owen      |  Harris     |  Braund   | Mr    |

Anything not clearly one of the above 4 categories can be ignored.

```python
df[['Last Name', 'Title', 'First Name', 'Middle Name']] = df['Name'].str.extract(r'([A-Za-z]+),\s([A-Za-z]+(?:\s[A-Za-z]+)*)\.\s*([A-Za-z]+)\s*(.*)')
df.head()
```

f) Complete the code below to replace all missing ages with the average age

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.head()
```

g) Plot a bar chart of the average age of those that survived and did not survive. Briefly comment on what you observe.

```python
import matplotlib.pyplot as plt
average_age_by_survival = df.groupby('Survived')['Age'].mean()
print(average_age_by_survival)

# Plotting the bar chart
average_age_by_survival.plot(kind='bar', color=['green', 'red'])
plt.title('Average Age by Survival')
plt.xlabel('Survived')
plt.ylabel('Average Age')
plt.xticks([0, 1], ['Survived', 'Not Survived'], rotation=0)
plt.show()
```
